```python
def routes_from_file(filename):
    with open("../routes/" + filename) as f:
        lines = f.readlines()
    return (int(lines[-1].split('\t')[0]),[int(item) for item in lines[-1].split('\t')[1].replace('\n','').split(',')])

def draw_img(filename,routes,locations=data['locations']):
    final_file = "images/VUT/" + (filename.replace('.route','-tsp.png'))
    draw_tsp_path.draw_routes(locations,routes,FINAL_IMAGE=final_file)
```

```python
def_val_dict = {
    'file' : "VUT",
    'iterations' : 100,
    'ants' : 50,
    'initial_pheromone' : 0.01,
    'q' : 0.5,
    'rho' : 0.03,
    'beta' : 4.0,
    'alpha' : 1.0,
    'elistist_val' : 0.2,
    'rank' : 1,
    }
systems = [
    #'rank',
    'simple',
    'elitist',
    #'maxmin',
    #'acs',
]

def calc_one_run(json,system):
    try:
        if system == 'elitist':
            system += " "+str(json['elistist_val'])
        elif system == 'rank':
            system += " "+str(json['rank'])
        out = os.system("cd ../libaco-master/acotsp/; ./bin/acotsp --{} -f benchmarks/darkVUTinverted-256-stipple.tsp -i {} -m {} -p {} -q {} -r {} -b {} -a {} > ../../routes/{}.route".format(
                system,
                json['iterations'],
                json['ants'],
                json['initial_pheromone'],
                json['q'],
                json['rho'],
                json['beta'],
                json['alpha'],
                json['file']
                )
            )
        if out != 0:
            print("some fucked up happend ",json['file'])
            return -1
        
        best_score, routes = routes_from_file(json['file']+'.route')
        print(json['file'],'with score',best_score,'saving....')
        draw_img(json['file']+'.route',routes)
        return best_score,routes
    except Exception as e:
        print(e)
        return -1
```

```python
def save_arr(name,arr):
    with open("images/VUT/"+name+".pkl", "wb") as saved_data:
        pickle.dump(arr, saved_data)
    saved_data.close()

def load_arr(name):
    with open("images/VUT/"+name+".pkl", "rb") as saved_data:
        return pickle.load(saved_data)
```

```python
runs = {}
base_name = '256darkVUTinverted'
#for system in systems:
#    index = 0
#    runs[system] = []
#    for i in [10,50,100,128,256]:
#        updated = def_val_dict.copy()
#        updated['file'] = base_name+'_'+system+'_ants_'+str(i)
#        updated['ants'] = i
#        index += 1
#        runs[system].append([calc_one_run(updated,system),updated])
#        save_arr(base_name+'_ants',runs)

for system in systems:
    runs[system] = []
    if system == 'elitist':
        for elit_val in [0.1,0.5,1.0,2.0,5.0]:
            updated = def_val_dict.copy()
            updated['file'] = base_name+'_'+system+'_elistist_val_'+str(elit_val)
            updated['elistist_val'] = elit_val
            runs[system].append([calc_one_run(updated,system),updated])
            save_arr(base_name+'_small_elitist',runs)

for system in systems:
    for beta in [0.1,1.0,2.0,5.0]:
        updated = def_val_dict.copy()
        updated['file'] = base_name+'_'+system+'_beta_'+str(beta)
        updated['alpha'] = 5.0 - beta
        updated['beta'] = beta
        runs[system].append([calc_one_run(updated,system),updated])
        save_arr(base_name+'_beta',runs)
    
for system in systems:
    for alpha in [0.1,1.0,2.0,5.0]:
        updated = def_val_dict.copy()
        updated['file'] = base_name+'_'+system+'_alpha_'+str(alpha)
        updated['beta'] = 5.0 - alpha
        updated['alpha'] = alpha
        runs[system].append([calc_one_run(updated,system),updated])
        save_arr(base_name+'_alpha',runs)

for system in systems:
    for rho in [0.01,0.1,0.5,1.0]:
        updated = def_val_dict.copy()
        updated['file'] = base_name+'_'+system+'_rho_'+str(rho)
        updated['rho'] = rho
        runs[system].append([calc_one_run(updated,system),updated])
        save_arr(base_name+'_rho',runs)

for system in systems:
    for q in [0.1,0.5,1.0,2.0]:
        updated = def_val_dict.copy()
        updated['file'] = base_name+'_'+system+'_q_'+str(q)
        updated['q'] = q
        runs[system].append([calc_one_run(updated,system),updated])
        save_arr(base_name+'_q',runs)

for system in systems:
    for pheromone in [-2.0,0.01,0.1,0.5,1.0]:
        updated = def_val_dict.copy()
        updated['file'] = base_name+'_'+system+'_pheromone_'+str(pheromone)
        updated['initial_pheromone'] = pheromone
        runs[system].append([calc_one_run(updated,system),updated])
        save_arr(base_name+'_pheromone',runs)
```
