```python
import pandas as pd
import pickle
```

```python
def save_arr(name,arr):
    with open("images/VUT/"+name+".pkl", "wb") as saved_data:
        pickle.dump(arr, saved_data)
    saved_data.close()

def load_arr(name):
    with open("images/VUT/"+name+".pkl", "rb") as saved_data:
        return pickle.load(saved_data)
```

```python
def transform_df(df):
    route = pd.DataFrame(df[0].tolist(),columns=['best','route'])
    json = pd.DataFrame(df[1].tolist())
    return pd.concat([route,json],axis=1)

ants = load_arr('256darkVUTinverted_ants')
runs = load_arr('256darkVUTinverted_pheromone')
dfS = transform_df(pd.DataFrame(runs['simple']+ants['simple']))
dfI = transform_df(pd.DataFrame(runs['elitist']+ants['elitist']))
```
